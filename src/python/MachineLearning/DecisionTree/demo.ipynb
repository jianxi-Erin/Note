```python
# demo1 决策树
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

# 数据
features = [[178, 1], [155, 0], [177, 0], [165, 0], [169, 1], [160, 1]]
labels = ['man', 'woman', 'man', 'woman', 'man', 'woman']

# 创建并训练决策树分类器
clf = DecisionTreeClassifier()
clf.fit(features, labels)

# 测试点
test_points = [[176, 0], [165, 1], [168, 1]]
predictions = clf.predict(test_points)
print("预测结果:", predictions)  # 输出：['man' 'woman' 'woman']

# 可视化决策树
# plt.figure(figsize=(10, 8))
plot_tree(clf, feature_names=['Height', 'Short Hair'], class_names=['man', 'woman'], filled=True, rounded=True)
plt.show()
```

```python
# demo2
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target)
# 基尼系数
clf_gini = DecisionTreeClassifier(criterion='gini')
clf_gini.fit(X_train, y_train)
print("Gini Train Score:", clf_gini.score(X_train, y_train))  # 约 1.0
print("Gini Test Score:", clf_gini.score(X_test, y_test))     # 约 0.95
plot_tree(clf_gini, 
          feature_names=iris.feature_names,  # 使用数据集的特征名称
          class_names=iris.target_names.tolist(),  # 使用数据集的类别名称 
          filled=True, rounded=True)
plt.show()
```

```python
# 信息熵
clf_entropy = DecisionTreeClassifier(criterion='entropy')
clf_entropy.fit(X_train, y_train)
print("Entropy Train Score:", clf_entropy.score(X_train, y_train))  # 约 1.0
print("Entropy Test Score:", clf_entropy.score(X_test, y_test))      # 约 0.92
plot_tree(clf_entropy, 
          feature_names=iris.feature_names,  # 使用数据集的特征名称
          class_names=iris.target_names.tolist(),  # 使用数据集的类别名称 
          filled=True, rounded=True)
plt.show()
```

```python

# demo3
# 限制生成树的深度
clf_depth3 = DecisionTreeClassifier(criterion='gini', max_depth=3)
clf_depth3.fit(X_train, y_train)
print("Max Depth=3 Test Score:", clf_depth3.score(X_test, y_test))  # 约 0.97
plot_tree(clf_depth3, 
          feature_names=iris.feature_names,  # 使用数据集的特征名称
          class_names=iris.target_names.tolist(),  # 使用数据集的类别名称 
          filled=True, rounded=True)
plt.show()
```
