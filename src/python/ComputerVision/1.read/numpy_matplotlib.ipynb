```python
import numpy as np  # 导入 numpy 库，用于数值计算和数组操作
import matplotlib.pyplot as plt  # 导入 matplotlib.pyplot 库，用于绘图
import cv2 # 导入opencv2库
plt.rcParams['font.sans-serif'] = ['SimHei']  # 使用黑体字体
```

```python
# numpy的简单使用
a=np.array((1,2,3),dtype=int)
print('a:',a)
print('类型:a.type:',a.dtype)
b=np.array([[1,2,3],[4,5,6]])
print('b:',b)
print('维度:b.ndim',b.ndim)
print('形状:b.shape',b.shape)
print("大小:b.size",b.size)
```

```python
# 散点图
x=np.array([1,2,3,4,5,6,7,8])
y=np.array([1,4,9,16,7,11,23,18])
plt.scatter(x,y,s=200,c='r')
plt.xlabel('x')
plt.ylabel('y')
plt.title('散点图')
plt.show()
```

```python
# 绘制折线图
# 创建 x 和 y 数据
x = np.array([0, 6])  # 定义 x 轴数据，包含两个点
y = np.array([0, 100])  # 定义 y 轴数据，与 x 轴数据对应，包含两个点

# 设置整体标题
plt.suptitle("网格标题")  # 为整个图形窗口设置一个总标题 "title"

# 使用 plt.subplot() 创建一个 2x2 的子图网格，并选择第一个子图进行绘制
plt.subplot(2, 2, 1)  # 参数 (2, 2, 1) 指定图像网格为 2 行 2 列，并选择第一个子图（左上角）

# 绘制 x 和 y 数据的折线图
plt.plot(x, y)  # 使用 plot() 函数绘制折线图

# 添加第一个子图的标题
plt.title("子图1")  # 为当前子图（第一个子图）添加标题 "p1"

# 创建第二组数据
x1 = np.array([1, 2, 3, 4])  # 定义第二组 x 轴数据
y1 = np.array([1, 4, 9, 16])  # 定义第二组 y 轴数据，与 x1 轴数据对应

# 选择第二个子图进行绘制
plt.subplot(2, 2, 2)  # 参数 (2, 2, 2) 指定图像网格的第二个子图（右上角）

# 绘制 x1 和 y1 数据的折线图
plt.plot(x1, y1)  # 使用 plot() 函数绘制折线图

# 添加第二个子图的标题
plt.title("子图2")  # 为当前子图（第二个子图）添加标题 "p2"

# 显示图像
plt.show()  # 显示所有子图和整体标题
```

```python


# 创建一个 800x500 像素的黑色图像，图像有 3 个通道（RGB），数据类型为 np.uint8
# np.zeros() 函数用于创建一个全为零的数组，在这里每个像素的值都初始化为 0 (黑色)
# 图像大小为 800x500，3 个通道代表 RGB，每个通道的值范围为 0-255
img = np.zeros((800, 500, 3), np.uint8)

# 输出图像的形状（宽度、高度、通道数）
print(img.shape)

# 绘制一条蓝色直线，起点为(20, 700)，终点为(200, 700)，线条宽度为5
cv2.line(img, (20, 700), (200, 700), (255, 0, 0), 5)

# 绘制一个矩形，左上角坐标为(20, 600)，右下角坐标为(200, 650)，颜色为黄色，线条宽度为3
cv2.rectangle(img, (20, 600), (200, 650), (255, 200, 0), 3)

# 绘制一个填充的黄色圆形，圆心坐标为(300, 300)，半径为50，-1表示填充
cv2.circle(img, (300, 300), 50, (255, 200, 0), -1)

# 绘制一个椭圆，圆心坐标为(300, 600)，长轴半径为100，短轴半径为50，逆时针旋转90度，起始角度为0度，结束角度为270度，颜色为黄色，线条宽度为3
cv2.ellipse(img, (300, 600), (100, 50), 90, 0, 270, (255, 200, 0), 3)

# 在图像上绘制文字 "paimeng"，位置为(0, 750)，字体为 FONT_HERSHEY_SIMPLEX，字体大小为1，颜色为黄色，线条宽度为2
cv2.putText(img, "paimeng", (0, 750), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 200, 0), 2)

# 绘制一个多边形，顶点为 [[50, 100], [150, 200], [300, 100], [200, 300], [50, 300]]，设置闭合，颜色为黄色，线条宽度为3
cv2.polylines(img, [np.array([[50, 100], [150, 200], [300, 100], [200, 300], [50, 300]])], True, (255, 200, 0), 3)

# 显示图像，窗口标题为 "title"
cv2.imshow("title", img)

# 等待按键按下
cv2.waitKey(0)

# 关闭所有窗口
cv2.destroyAllWindows()
```
